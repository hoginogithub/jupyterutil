```python
with open(household_budget_path, "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    csvlist = list(reader)
```

```python
# 昼食 玉子屋弁当　202107分のマッチング
total_fee = 0
item_name = "昼食 玉子屋弁当"
sum_if = "202107" 
for csv_item in csvlist:
    try:
        if item_name in csv_item[1] and sum_if in csv_item[4]:
            print(csv_item)
            total_fee += int(csv_item[2])
    except Exception:
        pass
print(f"{item_name=} {sum_if}分：{total_fee:,}円")
```
